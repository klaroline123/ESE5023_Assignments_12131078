```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import netCDF4 as nc
import xarray as xr

# question_1
ds = xr.open_dataset('./pcr_sm_1980_2014_monthly_050d.nc4', engine="netcdf4")

ds.sm.mean(dim=('lon', 'lat')).sel(time=slice("1980-01", "2017-06")).plot()
plt.show()
p_clim = ds.sm.groupby('time.season').mean()
print(p_clim)
p_clim.sel(season = 'DJF').plot()
plt.show()
p_clim.sel(season = 'JJA').plot()
plt.show()
p_clim.sel(season = 'MAM').plot()
plt.show()
p_clim.sel(season = 'SON').plot()
plt.show()
```
